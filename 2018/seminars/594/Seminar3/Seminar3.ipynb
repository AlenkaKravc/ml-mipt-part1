```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

### Загрузим данные

```python
df = pd.read_csv("startup_funding.csv")
df.drop(["SubVertical", "InvestorsName", "Remarks"], axis=1, inplace=1)
df.head()
```

```python
print(df.shape)

for col in df.columns:
    print(col, df[col].nunique(), df[col].isnull().any())
```

```python
df.Date = df.Date.str.replace(".", "/")

df["day"] = df.Date.str.split("/").str.get(0).astype(int)
df["month"] = df.Date.str.split("/").str.get(1).astype(int)
df["year"] = df.Date.str.split("/").str.get(-1).astype(int)

del df["Date"]
```

```python
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder, LabelBinarizer

to_encode = ["StartupName", "IndustryVertical", "CityLocation", "InvestmentType"]

for column in to_encode:
    le = LabelEncoder()
    df[column] = le.fit_transform(df[column])
```

```python
df.head()
```

```python
from sklearn.preprocessing import Imputer

pd.options.display.float_format = '{:.2f}'.format

df["AmountInUSD"] = df["AmountInUSD"].str.replace(",", "").astype(float)
df.fillna(0, inplace=True)

im = Imputer(missing_values=0)
df["AmountInUSD"] = im.fit_transform(df["AmountInUSD"].values.reshape(-1, 1))[:, 0]
df.head()
```
